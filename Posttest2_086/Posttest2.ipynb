## Hitung nilai unik

### Subtask:
Hitung dan tampilkan jumlah nilai unik untuk setiap kolom bertipe objek.


**Reasoning**:
Select object type columns and calculate the number of unique values for each.



```python
object_cols = df.select_dtypes(include='object').columns

print("Jumlah nilai unik untuk setiap kolom objek:")
for col in object_cols:
    unique_count = df[col].nunique()
    print(f"- {col}: {unique_count}")
```

## Visualisasi data

### Subtask:
Visualisasikan data menggunakan grafik yang sesuai.


**Reasoning**:
Generate count plots for categorical columns and histograms for numerical columns as requested in the instructions. Group related plots for better visualization.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set figure size for better readability
plt.figure(figsize=(15, 10))

# Plot count of Gender
plt.subplot(2, 3, 1)
sns.countplot(data=df, x='Gender')
plt.title('Count of Gender')

# Plot count of Sleep_Quality
plt.subplot(2, 3, 2)
sns.countplot(data=df, x='Sleep_Quality')
plt.title('Count of Sleep Quality')

# Plot count of Stress_Level
plt.subplot(2, 3, 3)
sns.countplot(data=df, x='Stress_Level')
plt.title('Count of Stress Level')

# Plot count of Health_Issues
plt.subplot(2, 3, 4)
sns.countplot(data=df, x='Health_Issues')
plt.title('Count of Health Issues')

# Plot count of Smoking
plt.subplot(2, 3, 5)
sns.countplot(data=df, x='Smoking')
plt.title('Count of Smoking')

# Plot count of Alcohol_Consumption
plt.subplot(2, 3, 6)
sns.countplot(data=df, x='Alcohol_Consumption')
plt.title('Count of Alcohol Consumption')

plt.tight_layout()
plt.show()

# Plot count of Country
plt.figure(figsize=(15, 7))
sns.countplot(data=df, x='Country')
plt.title('Count of Country')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()

# Plot count of Occupation
plt.figure(figsize=(15, 7))
sns.countplot(data=df, x='Occupation')
plt.title('Count of Occupation')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()

# Set figure size for histograms
plt.figure(figsize=(15, 10))

# Plot histogram of Age
plt.subplot(2, 3, 1)
sns.histplot(data=df, x='Age', kde=True)
plt.title('Distribution of Age')

# Plot histogram of Coffee_Intake
plt.subplot(2, 3, 2)
sns.histplot(data=df, x='Coffee_Intake', kde=True)
plt.title('Distribution of Coffee Intake')

# Plot histogram of Caffeine_mg
plt.subplot(2, 3, 3)
sns.histplot(data=df, x='Caffeine_mg', kde=True)
plt.title('Distribution of Caffeine mg')

# Plot histogram of Sleep_Hours
plt.subplot(2, 3, 4)
sns.histplot(data=df, x='Sleep_Hours', kde=True)
plt.title('Distribution of Sleep Hours')

# Plot histogram of BMI
plt.subplot(2, 3, 5)
sns.histplot(data=df, x='BMI', kde=True)
plt.title('Distribution of BMI')

# Plot histogram of Heart_Rate
plt.subplot(2, 3, 6)
sns.histplot(data=df, x='Heart_Rate', kde=True)
plt.title('Distribution of Heart Rate')

plt.tight_layout()
plt.show()

# Plot histogram of Physical_Activity_Hours
plt.figure(figsize=(8, 6))
sns.histplot(data=df, x='Physical_Activity_Hours', kde=True)
plt.title('Distribution of Physical Activity Hours')
plt.tight_layout()
plt.show()
```

## Korelasi

### Subtask:
Hitung dan tampilkan matriks korelasi menggunakan heatmap.


**Reasoning**:
Select only the numerical columns, calculate the correlation matrix, and create a heatmap to visualize the correlations.



```python
numerical_df = df.select_dtypes(include=['number'])
correlation_matrix = numerical_df.corr()

plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Heatmap Korelasi Antar Kolom Numerik')
plt.show()
```
