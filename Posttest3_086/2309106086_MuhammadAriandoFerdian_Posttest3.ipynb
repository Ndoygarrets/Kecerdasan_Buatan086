Setelah menangani nilai yang hilang, langkah selanjutnya adalah mengidentifikasi dan menangani data duplikat dalam DataFrame. Memeriksa duplikat akan membantu memastikan setiap entri data adalah unik.



```python
# Identifikasi data duplikat
jumlah_duplikat = df.duplicated().sum()

print(f"\nJumlah baris duplikat sebelum penanganan: {jumlah_duplikat}")

# Menangani data duplikat (jika ada) dengan menghapusnya
if jumlah_duplikat > 0:
    df.drop_duplicates(inplace=True)
    print(f"Jumlah baris setelah menghapus duplikat: {len(df)}")
else:
    print("Tidak ada data duplikat yang ditemukan.")
```

**Reasoning**:
Karena tidak ada nilai yang hilang dan tidak ada duplikat, langkah selanjutnya adalah mengidentifikasi dan menangani outlier pada kolom numerik yang relevan menggunakan metode IQR. Kolom seperti 'Age', 'Coffee_Intake', 'Caffeine_mg', 'Sleep_Hours', 'Sleep_Quality', 'BMI', 'Heart_Rate', 'Stress_Level', dan 'Physical_Activity_Hours' kemungkinan memiliki outlier yang perlu ditangani.



```python
import numpy as np

# Identifikasi kolom numerik yang relevan
numerical_cols = ['Age', 'Coffee_Intake', 'Caffeine_mg', 'Sleep_Hours', 'Sleep_Quality', 'BMI', 'Heart_Rate', 'Stress_Level', 'Physical_Activity_Hours']

print("Jumlah outlier per kolom (menggunakan metode IQR sebelum penanganan):")

# Menangani outlier menggunakan metode IQR dan capping
for col in numerical_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers_count = df[(df[col] < lower_bound) | (df[col] > upper_bound)].shape[0]
    print(f"- {col}: {outliers_count}")

    # Capping outlier
    df[col] = np.where(df[col] < lower_bound, lower_bound, df[col])
    df[col] = np.where(df[col] > upper_bound, upper_bound, df[col])

print("\nJumlah outlier per kolom (menggunakan metode IQR setelah penanganan):")
for col in numerical_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers_count = df[(df[col] < lower_bound) | (df[col] > upper_bound)].shape[0]
    print(f"- {col}: {outliers_count}")
```

**Pembaruan**: Kesalahan tipe data saat menangani outlier pada sel `2564ad0c` telah diperbaiki di sel kode selanjutnya (`rmhPK895N8Yk`) dengan memastikan operasi hanya diterapkan pada kolom numerik.

**Reasoning**:
Error traceback menunjukkan bahwa operasi kuantil tidak dapat dilakukan pada kolom string. Ini berarti kolom 'Sleep_Quality' atau 'Stress_Level' yang awalnya diasumsikan numerik ternyata mengandung nilai string. Perlu memeriksa tipe data kolom-kolom ini dan mengecualikan kolom non-numerik dari daftar `numerical_cols` sebelum mencoba menghitung kuantil.



```python
# Periksa tipe data dari kolom yang awalnya diasumsikan numerik
print("\nTipe data kolom numerik yang relevan:")
print(df[numerical_cols].dtypes)

# Perbarui daftar kolom numerik untuk mengecualikan kolom non-numerik
numerical_cols = ['Age', 'Coffee_Intake', 'Caffeine_mg', 'Sleep_Hours', 'BMI', 'Heart_Rate', 'Physical_Activity_Hours']

print("\nJumlah outlier per kolom (menggunakan metode IQR sebelum penanganan):")

# Menangani outlier menggunakan metode IQR dan capping
for col in numerical_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers_count = df[(df[col] < lower_bound) | (df[col] > upper_bound)].shape[0]
    print(f"- {col}: {outliers_count}")

    # Capping outlier
    df[col] = np.where(df[col] < lower_bound, lower_bound, df[col])
    df[col] = np.where(df[col] > upper_bound, upper_bound, df[col])

print("\nJumlah outlier per kolom (menggunakan metode IQR setelah penanganan):")
for col in numerical_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers_count = df[(df[col] < lower_bound) | (df[col] > upper_bound)].shape[0]
    print(f"- {col}: {outliers_count}")
```

## Feature engineering

### Subtask:
Membuat minimal satu fitur baru dari data yang ada.


**Reasoning**:
Membuat fitur baru `Coffee_Per_Sleep_Hour` dengan membagi kolom `Coffee_Intake` dengan `Sleep_Hours`, menangani potensi pembagian dengan nol, dan kemudian menampilkan beberapa baris pertama DataFrame untuk verifikasi.



```python
# Buat fitur baru 'Coffee_Per_Sleep_Hour'
# Tangani pembagian dengan nol: jika Sleep_Hours adalah 0 atau mendekati nol, tetapkan nilai tertentu (misalnya, 0 atau NaN, atau nilai besar)
# Mengingat Sleep_Hours telah distandarisasi, nilai 0 asli mungkin menjadi nilai yang sangat kecil mendekati rata-rata yang distandarisasi.
# Namun, untuk amannya, kita akan menangani nilai yang sangat kecil setelah standarisasi sebagai pembagi nol.
# Kita bisa mengganti nilai Sleep_Hours yang mendekati nol dengan nilai yang sangat kecil untuk menghindari inf, atau mengganti hasilnya dengan NaN.
# Mari kita ganti hasil pembagian dengan NaN jika Sleep_Hours sangat mendekati nol.
# Threshold 1e-9 dipilih secara arbitrari, bisa disesuaikan jika perlu.
threshold = 1e-9
df['Coffee_Per_Sleep_Hour'] = df['Coffee_Intake'] / np.where(np.abs(df['Sleep_Hours']) < threshold, np.nan, df['Sleep_Hours'])

# Tampilkan lima baris pertama dari DataFrame df untuk memverifikasi penambahan fitur baru
display(df.head())
```
